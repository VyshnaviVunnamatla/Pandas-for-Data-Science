# GETTING FAMILIAR WITH PANDAS

**Creating DataFrames and Series:**

creating from list

```python
import pandas as pd
data = [1, 2, 3, 4, 5]
series = pd.Series(data)
```

creating from dictionaries

```python
data = {'A': [1, 2, 3], 'B': [4, 5, 6]}
df = pd.DataFrame(data)
```

creating from csv file

```python
#df = pd.read_csv('data.csv')
```

**Basic Operations:**

```python
# Selecting Data

df['A']  # Selects column A
df.loc[0]  # Selects first row
```

```python
# Filtering Rows

df[df['A'] > 2]  # Filters rows where column A > 2
```

```python
# Modifying Data

df['A'] = df['A'] * 2  # Doubles column A
df['A']
```

# DATA HANDLING WITH PANDAS

**Reading Data from Files**

```python
#Reading CSV files:

#df = pd.read_csv('data.csv')

#Reading Excel files:

#df = pd.read_excel('data.xlsx')
```

**Handling Missing Data**

```python
# Detect missing data

df.isnull()  # Returns True for missing values
```

```python
# Fill missing data

df.fillna(0)  # Fills missing values with 0
```

```python
# Drop missing data

df.dropna()  # Drops rows with missing values
```

**Data Transformation**

```python
# Remove duplicates

df.drop_duplicates()  # Removes duplicate rows
```

```python
# Data type conversion

df['A'] = df['A'].astype(float)  # Converts column A to float
df['A']
```

# DATA ANALYSIS WITH PANDAS


**Generating Summary Statistics**

```python
df.describe()
```

**Grouping Data**

```python
grouped = df.groupby('A').sum()
print(grouped)
```

**Applying Aggregate Functions**

```python
df.agg({'A': 'mean', 'B': 'sum'})
```

**Advanced Data Manipulation**

```python
# Merging DataFrames:

df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['B', 'C', 'D'], 'value': [4, 5, 6]})

merged_df = pd.merge(df1, df2, on='key')
print(merged_df)
```

```python
# joining DataFrames:

joined_df = df1.join(df2, lsuffix='_left', rsuffix='_right')
print(joined_df)
```

```python
# Concatenating DataFrames:

concatenated_df = pd.concat([df1, df2])
print(concatenated_df)
```
